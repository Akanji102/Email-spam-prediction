```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
spam_prediction = pd.read_csv("spamhamdata.csv", sep="\t", header=None, names=["Category", "Message"])
```

```python
spam_prediction.head()
```

```python
spam_prediction.isnull().sum()
```

```python
spam_prediction = spam_prediction.where((pd.notnull(spam_prediction)), '')
```

```python
spam_prediction.head()
```

```python
#Checking the number of rows and columns in the ddtaset
spam_prediction.shape
```

Label Encoding

Spam mail as 0

ham mail as 1

```python
spam_prediction.loc[spam_prediction['Category'] == 'spam', 'Category',] = 0
spam_prediction.loc[spam_prediction['Category'] == 'ham', 'Category',] = 1
```

```python
spam_prediction.head()
```

```python
#Separating the data as tests and labels

x = spam_prediction['Message']
y = spam_prediction['Category']
```

```python
print(x)
```

```python
print(y)
```

Splitting the data into training data and test data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size = 0.2, random_state = 42)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

Feature Extraction

```python
#Transform the test data to feature vectors that can be used as input in the logistic regression

feature_extraction = TfidfVectorizer(min_df = 1, stop_words = 'english', lowercase = True)
```

```python
x_train_features = feature_extraction.fit_transform(x_train)
x_test_features = feature_extraction.transform(x_test)

#Convert y_train and y_test as int
y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
print(x_train)
```

```python
print(x_train_features)
```

Training the model

Logistic Regression

```python
model = LogisticRegression()
```

```python
#Training the logistic regression with the training data
model.fit(x_train_features, y_train)
```

Evaluating the trained model

```python
#prediction on training data

prediction_on_training_data = model.predict(x_train_features)
accuracy_on_training_data = accuracy_score(y_train, prediction_on_training_data)
```

```python
print('Accuracy on training data: ', accuracy_on_training_data)
```

```python
#prediction on test data

prediction_on_test_data = model.predict(x_test_features)
accuracy_on_test_data = accuracy_score(y_test, prediction_on_test_data)
```

```python
print('Accuracy on test data: ', accuracy_on_test_data)
```

```python
#Building a predictive system

input_data = ["WINNER!! As a valued network customer you have been selected to receivea £900 prize reward! To claim call 09061701461. Claim code KL341. Valid 12 hours only."]

#Convert data to feature vectors

input_data_features = feature_extraction.transform(input_data)

#Making predictions

predictions = model.predict(input_data_features)
print(predictions)

if (predictions[0] == 1):
    print("I am 96% sure that this is a ham mail")
else:
    print("I am very sure this is a spam mail")
```
